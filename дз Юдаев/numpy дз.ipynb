```python
# Задача 2
import numpy as np
import imageio
import matplotlib.pyplot as plt
# Загрузка изображения
img = imageio.imread("test.png")
# Коэффициенты для преобразования в оттенки серого
coeffs = np.array([0.299, 0.587, 0.114, 0.114])

# Преобразование в оттенки серого
gray_img = np.dot(img, coeffs)

# Веса каналов
weights = np.array([0.3, 0.5, 0.2, 0.2])

# Сложение каналов с указанными весами
result = np.sum(img * weights, axis=2)

# Вывод результата
print(result.shape) # (height, width)


# Показать результирующее изображение в градациях серого
plt.imshow(gray_img, cmap='gray')
plt.show()

# Показать результирующее изображение взвешенной суммы
plt.imshow(result, cmap='gray')
plt.show()
```

```python
# Задача 3

import numpy as np
from scipy.linalg import solve
import matplotlib.pyplot as plt

# Задаем функцию f(x)
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)

# точки, через которые должен проходить многочлен
x1 = 1
x2 = 15

# система линейных уравнений для многочлена 1й степени
A = np.array([[1, x1], [1, x2]])
b = np.array([f(x1), f(x2)])
w1 = solve(A, b)

# Формируем систему линейных уравнений для многочлена второй степени
x3 = 8
A = np.array([[1, x1, x1**2], [1, x3, x3**2], [1, x2, x2**2]])
b = np.array([f(x1), f(x3), f(x2)])
w2 = solve(A, b)

# Формируем систему линейных уравнений для многочлена третьей степени
x4 = 4
x5 = 10
A = np.array([[1, x1, x1**2, x1**3], [1, x4, x4**2, x4**3], [1, x5, x5**2, x5**3], [1, x2, x2**2, x2**3]])
b = np.array([f(x1), f(x4), f(x5), f(x2)])
w3 = solve(A, b)

# Выводим коэффициенты многочленов
print('Коэффициенты многочлена 1й степени:',w1)
print('Коэффициенты многочлена 2й степени:',w2)
print('Коэффициенты многочлена 3й степени:',w3)

# Рисуем графики функции f(x) и многочленов первой, второй и третьей степени
x = np.linspace(1, 15, 100)
plt.plot(x, f(x), label='f(x)')
plt.plot(x, w1[0] + w1[1]*x, label='1я степень')
plt.plot(x, w2[0] + w2[1]*x + w2[2]*x**2, label='2я степень')
plt.plot(x, w3[0] + w3[1]*x + w3[2]*x**2 + w3[3]*x**3, label='3я степень')
plt.legend()
plt.show()
```

```python
#На графике видно, что многочлен первой и второй степени плохо приближает функцию f(x),
#а многочлен третьей степени даёт более точное приближение. 
```
